# 4) Estructuras de control: `if`, `for`, `while`

## ¿Qué son?
Mecanismos para **tomar decisiones** (`if`) y **repetir acciones** (`for`, `while`).

## ¿Para qué se usan?
Para **controlar el flujo** de un programa según condiciones y colecciones.

## Ideas clave
- La **indentación** define bloques.
- `for` itera sobre secuencias; `range(n)` genera 0..n-1.
- `while` repite mientras la condición sea verdadera.
- `break`/`continue` controlan el bucle.


```python
# Par o impar
n = 7
if n % 2 == 0:
    print("Par")
else:
    print("Impar")

# Suma 1..10
total = 0
for i in range(1, 11):
    total += i
print("Suma 1..10:", total)

# Cuenta regresiva con while
c = 5
while c > 0:
    print(c)
    c -= 1
```

### Ejercicios
- E1: Clasifica una nota (0–100) en A/B/C/D/F.
- E2: Imprime la tabla de multiplicar de un número (1–10).
- E3: Con `while`, pide números hasta 0; muestra suma y promedio.

### Reto extra
**FizzBuzz** de 1 a 100.


```python
nota = float(input("Ingrese la nota (0-100): "))

if nota < 0 or nota > 100:
    print("Nota inválida")
elif nota >= 90:
    print("A - Excelente")
elif nota >= 80:
    print("B - Bueno")
elif nota >= 70:
    print("C - Regular")
elif nota >= 60:
    print("D - Mala nota")
else:
    print("F - Nota 0")
```

```python
print("Tabla de multiplicar del 7:")
for i in range(1, 11):
    resultado = 7 * i
    print(f"7 multiplicado por {i} es igual a {resultado}")
```

```python
suma = 0
cuantos = 0

while True:
    n = float(input("Ingrese un número (0 para terminar): "))
    if n == 0: break
    suma = suma + n
    cuantos = cuantos + 1

if cuantos > 0:
    print("Suma:", suma)
    print("Promedio:", suma / cuantos)
else:
    print("No hay números")
```

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```
